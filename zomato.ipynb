```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\dholi\\OneDrive\\Desktop\\datasets\\zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if str(value).strip().lower() in ['new', '-']:
        return np.nan
    elif '/' in str(value):
        parts = str(value).split('/')
        return float(parts[0])
    else:
        try:
            return float(value)
        except ValueError:
            return np.nan
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates']=df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handle_rest_type) 
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts(ascending  = False)
location_lessthan300 = location[location<300]
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines_lessthan100 = cuisines[cuisines<100]
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
# Distribution of restaurant ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='rate', data=df, order=df['rate'].value_counts().index)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Top cuisines in the dataset
top_cuisines = df['cuisines'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_cuisines.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine Type')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Location-wise distribution of restaurants
plt.figure(figsize=(14, 8))
sns.countplot(x='location', data=df, order=df['location'].value_counts().head(15).index)
plt.title('Location-wise Distribution of Restaurants')
plt.xlabel('Location')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Distribution of average cost for two
plt.figure(figsize=(10, 6))
sns.histplot(df['Cost2plates'], bins=30, kde=True, color='orange')
plt.title('Distribution of Cost2plates')
plt.xlabel('Cost2plates')
plt.ylabel('Frequency')
plt.show()
```

```python
# Types of restaurants
plt.figure(figsize=(10, 6))
sns.countplot(x='rest_type', data=df, order=df['rest_type'].value_counts().head(10).index)
plt.title('Top 10 Restaurant Types')
plt.xlabel('Restaurant Type')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.show()
```
